```python
import pandas as pd
```

```python
review_df = pd.read_csv('output/reviews_scores.csv')
```

```python
review_df.head()
```

```python

grouped = review_df.groupby(['restaurant_ids', 'name', 'cuisine']).size().reset_index(name='count')

# Pivot the table to have one row per restaurant and output counts as columns
review_agg_df = grouped.pivot_table(index=['restaurant_ids', 'name', 'cuisine'], values='count', fill_value=0).reset_index()

# Calculate total reviews per restaurant
review_agg_df['total'] = review_agg_df['positive'] + review_agg_df['negative'] + review_agg_df['neutral']

# Calculate percentages
review_agg_df['positive_percentage'] = review_agg_df['positive'] / review_agg_df['total'] * 100
review_agg_df['negative_percentage'] = review_agg_df['negative'] / review_agg_df['total'] * 100
review_agg_df['neutral_percentage'] = review_agg_df['neutral'] / review_agg_df['total'] * 100

# Rename columns for better readability
review_agg_df.rename(columns={'positive': 'positive_count', 'negative': 'negative_count', 'neutral': 'neutral_count'}, inplace=True)

# Remove the 'total' column as it is not needed in the final output
# review_agg_df.drop(columns=['output'], inplace=True)

review_agg_df.head()



    
```
