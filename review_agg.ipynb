```python
import pandas as pd
```

```python
review_df = pd.read_csv('output/reviews_scores.csv')
```

```python
review_df.head()
```

```python

grouped = review_df.groupby(['restaurant_ids', 'name', 'cuisine', 'output']).size().reset_index(name='count')

# Pivot the table to have one row per restaurant and output counts as columns
review_agg_df = grouped.pivot_table(index=['restaurant_ids', 'name', 'cuisine'], columns='output', values='count', fill_value=0).reset_index()

# Calculate total reviews per restaurant
review_agg_df['total'] = review_agg_df['positive'] + review_agg_df['negative'] + review_agg_df['neutral']

# Calculate percentages
review_agg_df['positive_percentage'] = (review_agg_df['positive'] / review_agg_df['total'] * 100).round(0)
review_agg_df['negative_percentage'] = (review_agg_df['negative'] / review_agg_df['total'] * 100).round(0)
review_agg_df['neutral_percentage'] = (review_agg_df['neutral'] / review_agg_df['total'] * 100).round(0)

# Rename columns for better readability
review_agg_df.rename(columns={'positive': 'positive_count', 'negative': 'negative_count', 'neutral': 'neutral_count'}, inplace=True)


review_agg_df.head()
```

```python
# Convert percentages to type int
review_agg_df[['positive_percentage','negative_percentage','neutral_percentage']] = review_agg_df[['positive_percentage','negative_percentage','neutral_percentage']].astype('int')
review_agg_df.head()
```

```python
# Save!
review_agg_df.to_csv('output/cleaned_review_aggs.csv', index=False)
```

```python
# # Group by username instead - Curious
user_grouped = review_df.groupby(['username', 'output']).size().reset_index(name='count')

user_review_agg_df = user_grouped.pivot_table(index='username', columns='output', values='count', fill_value=0).reset_index()

user_review_agg_df['total'] = user_review_agg_df['negative'] + user_review_agg_df['neutral'] + user_review_agg_df['positive']

user_review_agg_df = user_review_agg_df.sort_values(by='total', ascending=False)

user_review_agg_df.head(20)
```
